# 1. selenium

```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time

# 페이지 열기
driver = webdriver.Chrome('./chromedriver.exe')
driver.get('https://cafe.daum.net/astronomical')
time.sleep(0.5)

# 로그인 클릭
driver.switch_to.frame('down')
elem = driver.find_element_by_xpath('//*[@id="loginout"]')
elem.click()
time.sleep(0.5)

# 카카오계정 로그인
elem = driver.find_element_by_xpath('//*[@id="mArticle"]/div[1]/div/div/div[2]/a[1]/span[2]')
elem.click()
time.sleep(0.5)

# 아이디 및 비밀번호 입력
with open('C:/Users/dojun/Documents/api_keys\daum.txt') as f:
    id = f.readline().replace('\n', '')
    pw = f.readline()

box_id = driver.find_element_by_xpath('//*[@id="id_email_2"]')
box_id.send_keys(id)

box_pw = driver.find_element_by_xpath('//*[@id="id_password_3"]')
box_pw.send_keys(pw)
box_pw.send_keys(Keys.ENTER)
time.sleep(0.5)

# 최신글 보기
driver.switch_to.frame('down')
elem = driver.find_element_by_xpath('//*[@id="fldlink_recent_bbs"]')
elem.click()
time.sleep(0.5)

# html 수집
html = driver.page_source
soup = bs(html, 'html.parser')
div = soup.find_all('span', class_='link_board')
del div[0:4]

# 링크 수집
ls_url = []
for d in div:
    href = d.find('a')['href']
    ls_url.append(href)
```

```python
# 페이지 이동하며 html 수집
html = driver.page_source
soup = bs(html, 'html.parser')
div = soup.find_all('span', class_='link_board')
del div[0:4]

# 링크 수집
for d in div:
    href = d.find('a')['href']
    ls_url.append(href)

len(ls_url)
```

# *ls_url to data frame

```python
import pandas as pd

df = pd.DataFrame(ls_url)
df.to_csv('astronomy_url_list.csv')
```

# 3. 형태소 분석

```python
from ckonlpy.tag import Twitter, Postprocessor

twitter = Twitter()
twitter.add_dictionary(['', ''], 'Noun')

passtags = {'Noun'}
postprocessor = Postprocessor(twitter, passtags=passtags)
```

```python
tpl = postprocessor.pos(txt)
```

```python
import pandas as pd

df = pd.DataFrame(tpl)
df
```

```python
txt = ''
for word in df[0]:
    txt += ' ' + word
```

```python
with open('./astronomy_nouns.txt', 'w', encoding='utf-8') as f:
    f.write(txt)
```
