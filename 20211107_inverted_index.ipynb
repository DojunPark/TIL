```python
tmp = df[:10]
tmp
```

```python
import re

# 1. tokenization
postings = []
for i, row in tmp.iterrows():
    txt = row[3].lower()
    ls = re.findall("[\w']+", txt)
    postings.append(ls)
    
print(postings)
```

```python
# 2. tagging numbers

ls_terms = []
for i, plist in enumerate(postings):
    for term in plist:
        ls_terms.append([term, i])
        
print(ls_terms)
```

```python
# 3. sorting

import pandas as pd

terms = pd.DataFrame(ls_terms)
terms.sort_values(by=[0, 1], inplace=True, ignore_index=True)
terms.head()
```

```python
# delete duplicate data

tmp = []
for i, t in terms.iterrows():
    if (t[0], t[1]) not in tmp:
        tmp.append((t[0], t[1]))

print(len(tmp))
print(tmp)
```
